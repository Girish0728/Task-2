```python
stock_prices = {
    "AAPL": 180,
    "TSLA": 250,
    "GOOG": 140,
    "TATA": 121,
    "NIFTY": 276,
    "GOLD": 627,
    "USD": 93
}

total = 0

while True:
    stock = input("Enter stock name (or Enter 'E' for exist): ").upper()
    
    if stock == "E":
        break
    
    if stock in stock_prices:
        quantity = int(input("Enter quantity: "))
        value = stock_prices[stock] * quantity
        total += value
        print("Value:", value)
    else:
        print("Stock not found")

print("Total Investment:", total)
```
